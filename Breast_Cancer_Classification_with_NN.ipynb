```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_data = sklearn.datasets.load_breast_cancer()
print(breast_cancer_data)
```

```python
data = pd.DataFrame(breast_cancer_data.data,columns = breast_cancer_data.feature_names)
data
```

```python
data['label'] = breast_cancer_data.target
data.head()
```

```python
X = data.drop(columns='label', axis=1)
Y = data['label']
```

## Splitting the data into training data & testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

### Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

### Building the nural networks

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of nural networks

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),                 # input layer
                          keras.layers.Dense(20,activation='relu'),                # hidden layer
                          # keras.layers.Dense(30,activation='sigmoid'),           # hidden layer - we can add more hidden layers
                          keras.layers.Dense(2,activation='sigmoid')               # output layer
      ])
```

```python
# Compiling the nural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# one hot encoding - categorical_crossentropy    (not sparse_categorical_crossentropy)
# sparse_categorical_crossentropy - for binary classification
```

```python
# training the nural network

history = model.fit(X_train_std, Y_train, epochs=10,validation_split=0.1)
```

### Visualization accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='upper right')
```

### Accuracy of the model on test data

```python
loss,accuracy = model.evaluate(X_test_std,Y_test)
print(accuracy)
```
